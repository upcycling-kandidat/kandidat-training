```python
# Material Detection
#https://arxiv.org/pdf/2212.00648v4.pdf
import os
import subprocess
path = '/content/drive/MyDrive/AI_TESTS_UPCYCLING/Net_For_Material_Similarity_CODE_AND_TRAINED_MODEL/One_shot_materials_recognition_using_contrastive-learnig-material-similarty--main'
input_image = f'{path}/sample_data/test/Candies_candy_IMG_20220506_165103_792.jpg'
input_mask = f'{path}/sample_data/test/Candies_candy_IMG_20220506_165103_792_MASK.png'
input_trained = f'{path}/logs/Defult.torch'
script_path = f'{path}/Test_Predict_descriptor.py'

# Run the command
cmd = f'python {script_path} --input_image {input_image} --input_mask {input_mask} --train_model_path {input_trained}'

process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE, shell=True)
output, error = process.communicate()
len(output)

print("Output:", output)
print("Error:", error)
print("Exit code:", process.returncode)

#!python3 Test_Predict_descriptor.py --input_image sample_data/test/Candies_candy_IMG_20220506_165103_792.jpg --input_mask sample_data/test/Candies_candy_IMG_20220506_165103_792_MASK.png --train_model_path logs/Defult.torch
```
